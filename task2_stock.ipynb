```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = yf.download('AAPL', period='6mo')
data.head()
```

```python
X = data[['Open', 'High', 'Low', 'Volume']]
y = data['Close']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
plt.plot(y_test.values, label='Actual')
plt.plot(predictions, label='Predicted')
plt.legend()
plt.show()
```

This project predicts stock closing prices using Linear Regression.

The model learns from stock features such as Open, High, Low, and Volume to predict the Close price.

The graph shows comparison between actual and predicted values.
